```python
import pandas as pd
import re
```

```python
data=open("C:\\Users\\YOGENDRA SINGH\\Desktop\\conv.txt",mode='r+')
lines=data.readlines()
```

```python
lines
```

```python
name_word={}
for line in lines:
    temp=line.split(":")
    if temp[0]not in name_word and temp[0]!='\n':
        name_word[temp[0]]=[]
```

```python
name_word
```

```python
for line in lines:
    temp=line.split(":")
    words=temp[-1].split(" ")
    for word in words:
        if temp[0]!='\n':
            name_word[temp[0]].append(word.upper())
```

```python
for lst in name_word:
    name_word[lst]=list(set(name_word[lst]))
    name_word[lst]=[re.sub("[\n+.+,+' '+…+?]","",x)for x in name_word[lst]]
    name_word[lst]=list(filter(lambda x: (x!=''),name_word[lst]))
    name_word[lst]=list(filter(lambda x:not x.isnumeric(),name_word[lst]))
    file=open(f"{lst}.txt","w+")
    for wrd in name_word[lst]:
        file.write(wrd+"\n")
    file.close()
```
